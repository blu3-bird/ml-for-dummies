### Practice Exercise (Data Preprocessing)

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
df = pd.read_csv('employee_salary.csv')
print(df.shape)
print(df.isnull().sum())

df.head(1)
```

```python
#feature and target variable
X = df.drop(['EmpID','Salary'],axis=1).values
y= df['Salary'].values

df.head(1)
```

```python
#handing the missing values
impute_mean = SimpleImputer(missing_values=np.nan,strategy='mean')

X[:,[0,9,10]] = impute_mean.fit_transform(X[:,[0,9,10]])

impute_frequent = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
X[:,[1,2,3,4,5,6,7,8]] = impute_frequent.fit_transform(X[:,[1,2,3,4,5,6,7,8]])
```

```python
ct = ColumnTransformer(transformers=[('encode',OneHotEncoder(),[1,2,3,4,6,7,8])],remainder='passthrough')
X = np.array(ct.fit_transform(X))

```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
#scaler featuer
sc = StandardScaler()
X_train[:,-3:] = sc.fit_transform(X_train[:,-3:])

X_test[:,-3:] = sc.transform(X_test[:,-3:])
```
