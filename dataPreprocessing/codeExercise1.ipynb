### Data Preprocessing Exercies

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
```

```python
#loading the dataset
df = pd.read_csv('loan.csv')
print(f'Loading the dataset:\n')
df.head()
```

```python
# defining the feature and target variable
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
categorical_columns = ['Gender', 'Married', 'Education', 'PropertyArea']
numerical_columns = ['Income', 'LoanAmount', 'CreditScore']
```

```python
#handling the missing data
missing_data = df.isnull().sum()
print(f'Missing data in each column:\n {missing_data}')

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X[:,3:6] = imputer.fit_transform(X[:,3:6])

impute = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
X[:,0:1] = impute.fit_transform(X[:,0:1])


```

```python
#Encoding the categories feature variable
ct = ColumnTransformer(transformers=[('arson',OneHotEncoder(),[0,1,2,6])],remainder='passthrough')
X = np.array(ct.fit_transform(X))

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
#spliting the data
X_train, X_test , y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=7)
```

```python
#feature Scaler
sc = StandardScaler()
X_train[:,10:] = sc.fit_transform(X_train[:,10:])

X_test[:,10:] = sc.transform(X_test[:,10:])
```

```python

#printing the final result: 
print(f'X_train:\n {X_train}')
print(f'X_test:\n {X_test}')
print(f'y_train:\n {y_train}')
print(f'y_test: {y_test}')
```
