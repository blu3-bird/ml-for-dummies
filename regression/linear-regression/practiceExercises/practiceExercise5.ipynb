### Practice Exercise (Linear Regression)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#loading the dataset
df = pd.read_csv('delivery_time.csv')
print(df.shape)

X = df[['Distance_KM']].values
y = df.iloc[:,-1].values
df.head(1)
```

```python
#splitting the data
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=69)
```

```python
ln = LinearRegression()
ln.fit(X_train,y_train)
```

```python
#making the predictin
pred_y = ln.predict(X_test)
```

```python
#visualization
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.scatter(X_train,y_train,color='r')
plt.plot(X_train,ln.predict(X_train),color='blue')
plt.title("Delivery Time (Training Set)")
plt.ylabel("Delivery Time in Min")
plt.xlabel("Distance in KM")

plt.subplot(1,2,2)
plt.scatter(X_test,y_test, color='r')
plt.plot(X_train,ln.predict(X_train),color='blue')
plt.title("Delivery Time (Test Set)")
plt.ylabel("Delivery Time in Min")
plt.xlabel("Distance in KM")

plt.tight_layout()
plt.show()
```
