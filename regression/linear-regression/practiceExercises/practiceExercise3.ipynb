### Practice Exercise 3 (Linear Regression)

#### Car Price Vs Years old

```python
#import 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#loading the Dataset
df = pd.read_csv('car-price.csv')

X = df.drop(['Price_USD','Mileage_K'],axis=1).values
y = df.iloc[:,-1].values

```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
ln = LinearRegression()
ln.fit(X_train,y_train)
```

```python
print(f'Intercept Value: {ln.intercept_:.2f}')
print(f'Coefficent Value: {ln.coef_[0]:.2f}')
```

```python
#making the prediction
pred_y = ln.predict(X_test)
```

```python
#visualization of dataset 
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.scatter(X_train,y_train,color='Yellow')
plt.plot(X_train, ln.predict(X_train),color='red')
plt.title('Car price vs Year made(training Set)')
plt.xlabel('Year old')
plt.ylabel('Car Price')

plt.subplot(1,2,2)
plt.scatter(X_test,y_test,color='Yellow')
plt.plot(X_train,ln.predict(X_train),color='red')
plt.title("Car Price VS Year made (Test Set)")
plt.ylabel("Car Price")
plt.xlabel('Year old')

plt.tight_layout()
plt.show()
```

#### Car Price Vs Milage

```python
x = df.drop(['Price_USD','Age_Years'],axis=1).values
Y = df.iloc[:,-1]
```

```python
# spliting the sets
x_train , x_test, Y_train, Y_test = train_test_split(x,Y,test_size=0.3,random_state=4)
```

```python
ln_m = LinearRegression()
ln_m.fit(x_train, Y_train)
```

```python
pred_y_m = ln_m.predict(x_test)
```

```python
#visualization of dataset
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.scatter(x_train,Y_train,color='Yellow')
plt.plot(x_train,ln_m.predict(x_train),color='red')
plt.title("Car Price VS Milage (Training set)")
plt.xlabel("Milage")
plt.ylabel("Car Price")

plt.subplot(1,2,2)
plt.scatter(x_test, Y_test, color='Yellow')
plt.plot(x_train,ln_m.predict(x_train),color='red')
plt.title("Car Price VS Milage (Test Set)")
plt.xlabel("Milage")
plt.ylabel("Car Price")

plt.tight_layout()
plt.show()
```
