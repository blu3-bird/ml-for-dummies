## Linear Regression Practice Exercise 2

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#loading the dataset
df = pd.read_csv('house-prices.csv')
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
# spliting the training and test set
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
# linear regression ob
ln = LinearRegression()
ln.fit(X_train, y_train)

print(f'Intercept Value: {ln.intercept_:.2f}')
print(f'Coeficent value: {ln.coef_[0]:.2f}')
```

```python
#making the prediction
pred_y = ln.predict(X_test)
```

```python
#visualization of training set and test set
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.scatter(X_train, y_train, color='yellow')
plt.plot(X_train, ln.predict(X_train),color='blue')
plt.title("House Price vs size (Training Set)")
plt.xlabel("Size")
plt.ylabel('House Price')

plt.subplot(1,2,2)
plt.scatter(X_test, y_test, color='yellow')
plt.plot(X_train,ln.predict(X_train),color='blue')
plt.title("House Price Vs Size (Test Set)")
plt.xlabel("Size")
plt.ylabel("House Price")

plt.tight_layout()
plt.show()
```
