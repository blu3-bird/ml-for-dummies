### Linear Regresssion Practice Exercise

```python
#Imports
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# loading the dataset
df = pd.read_csv('student_study.csv')
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#splitting the training and test set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
# linear regression obj
ln = LinearRegression()
ln.fit(X_train,y_train)
```

```python
# loading the intercept and slope
print(f'Intercept Value: {ln.intercept_:.2f}')
print(f'Slope Value: {ln.coef_[0]:.2f}')
```

```python
# making the prediction
pred_y = ln.predict(X_test)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X_train,y_train,color='blue')
plt.plot(X_train,ln.predict(X_train),color='red')
plt.title('Student Study Hours vs Marks (Training Set)')
plt.ylabel('Marks')
plt.xlabel('Hours')

plt.subplot(1,2,2)
plt.scatter(X_test, y_test, color='green')
plt.plot(X_train,ln.predict(X_train),color='red')
plt.title('Students Study hours vs Marks(Test set)')
plt.ylabel('Marks')
plt.xlabel('Hours')

plt.tight_layout()
plt.show()
```
